```python
# Загружаем информацию о товарах
data_meta = pd.read_json("C:/Users/n3sm3/Desktop/РЕКОМЕНДАТЕЛЬНЫЕ СИСТЕМЫ/meta_data.json", lines= True)
```

```python
# Данные содержат разную информацию об отзывах, в том числе время и сам текст отзыва
# Возьмем только информацию о пользователях, товарах и оценках
data_reviews.head()
```

```python
# reviewerID - пользователь
# asin - товар
# overall - оценка
data_reviews = data_reviews[['reviewerID','asin','overall']]
data_reviews
```

```python
# Из метаданных выбираем то, что понадобится для характеристики товаров
data_meta.head()
```

```python
# asin - товар
# brand - бренд
# feature - характеристики
# rank - популярность (топ в разделе)
data_meta = data_meta[['asin','brand','feature','rank']]
data_meta
```

```python
# В feature находится много информации, возьмем несколько параметров, которые можно попробовать использовать
# Создадим функцию для определения материала 
def material_category(row):
    str_ = str(row['feature']).lower()
    if 'cotton' in str_:
        return 'cotton'
    elif 'polyester' in str_:
        return 'polyester'
    elif 'spandex' in str_:
        return 'spandex'
    elif 'steel' in str_:
        return 'steel'
    elif 'lycra' in str_:
        return 'lycra'
    elif 'elastane' in str_:
        return 'elastane'
    elif 'nylon' in str_:
        return 'nylon'
    elif 'wool' in str_:
        return 'wool'
    elif 'synthetic' in str_:
        return 'synthetic'
    else:
        return 'no information'
```

```python
# Создадим функцию для определения категории товара по полу
def gender_category(row):
    str_ = str(row['feature']).lower()
    if 'girl' in str_:
        return 'female'
    elif 'woman' in str_:
        return 'female'
    elif 'women' in str_:
        return 'female'
    elif 'boy' in str_:
        return 'male'
    elif 'man' in str_:
        return 'male'
    elif 'men' in str_:
        return 'male'
    else:
        return 'no information'
```

```python
# Определяем категорию популярности
def rank_category(row):
    str_ = str(row['rank']).split('i')
    s = str_[0]
    return s
```

```python
# Получаем новую информацию
data_meta['material_category'] = data_meta.apply(material_category, axis = 1)
data_meta['gender_category'] = data_meta.apply(gender_category, axis = 1)
data_meta['rank_category'] = data_meta.apply(rank_category, axis = 1)
data_meta
```
